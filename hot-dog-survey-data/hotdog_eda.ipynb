```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("hot_dog_survey_spring25.xlsx - Form Responses 1.csv")
```

```python
# clean up hotdog column 
df.iloc[:,1] = df.iloc[:,1].str.upper().str[0].replace({'Y':'Yes','N':'No'})
```

```python
# select only undergraduate students 
df = df.loc[df.iloc[:,3]=='Yes']
```

```python
df
```

```python
# pie chart
import matplotlib.pyplot as plt

data = pd.DataFrame(df.iloc[:,1].value_counts())

plt.figure(figsize=(6, 6))  # Set figure size
plt.pie(data['count'], labels=data.index, autopct='%1.1f%%', startangle=140, colors=['red', 'green'])

# Add title
plt.title('Is a hotdog a sandwich?')

# Show the chart
plt.show()
```

```python
# bar chart by year
data2 = df.groupby(['Do you believe that a hotdog falls under the category of a sandwich?', 'If you are an undergraduate, what is your current academic year?']).size().reset_index(name='Count')
df_pivot = data2.pivot(index='Do you believe that a hotdog falls under the category of a sandwich?', columns='If you are an undergraduate, what is your current academic year?', values='Count')

df_pivot.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.xlabel('Is a hotdog a sandwich?')
plt.ylabel('Count')
plt.title('Hotdog Opinion by Academic Year')
plt.legend(title='Academic Year')
plt.show()
```

```python
# bar chart by gender
data3 = df.groupby(['Do you believe that a hotdog falls under the category of a sandwich?', 'Gender: How do you identify?']).size().reset_index(name='Count')
df_pivot = data3.pivot(index='Do you believe that a hotdog falls under the category of a sandwich?', columns='Gender: How do you identify?', values='Count')

df_pivot.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.xlabel('Is a hotdog a sandwich?')
plt.ylabel('Count')
plt.title('Hotdog Opinion by Gender')
plt.legend(title='Gender')
plt.show()
```
